# XGBoost with the Ames Housing Dataset

### Using XGBoost with regression
_This notebook is designed to take cleaned data from the Ames, Iowa housing dataset and apply a linear regression model with XGBoost regression. Furthermore, there will be evaluation of the XGBoost._

```python
# Loading appropriate libraries
import xgboost
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from xgboost.sklearn import XGBRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)

%matplotlib inline
```

```python
# Loading in data from previously cleaned datasets
num_train = pd.read_csv('../datasets/num_train.csv')
cat_train = pd.read_csv('../datasets/cat_train.csv')
num_test = pd.read_csv('../datasets/num_test.csv')
cat_test = pd.read_csv('../datasets/cat_test.csv')
train = pd.read_csv('../datasets/train.csv')
test = pd.read_csv('../datasets/test.csv')

# From previous data analysis, there are the features that have the highest correlation
# to the target, SalePrice
X_num_train = num_train[['Overall Qual', 'Lot Area', 'Year Built', 'Year Remod/Add', 
                         'Mas Vnr Area','BsmtFin SF 1', 'Total Bsmt SF', '1st Flr SF', 
                         'Gr Liv Area', 'Garage Area', 'Wood Deck SF', 'Screen Porch', 
                         'Pool Area', 'Misc Val', 'Kitchen Qual', 'Exter Qual', 'Bsmt Qual',
                         'Garage Cars', 'Garage Finish']]

y = train['SalePrice']

# Taking the same features for the final test data
X_num_test = num_test [['Overall Qual', 'Lot Area', 'Year Built', 'Year Remod/Add', 
                        'Mas Vnr Area','Total Bsmt SF', 'BsmtFin SF 1', '1st Flr SF', 
                        'Gr Liv Area', 'Garage Area', 'Wood Deck SF', 'Screen Porch', 
                        'Pool Area', 'Misc Val', 'Kitchen Qual', 'Exter Qual', 'Bsmt Qual',
                        'Garage Cars', 'Garage Finish']]

# Initializing Standard Scaler to standarize all the numerical features
ss = StandardScaler()
X_nums = pd.DataFrame(ss.fit_transform(X_num_train), columns = X_num_train.columns)
X_nums_final = pd.DataFrame(ss.transform(X_num_test), columns = X_num_test.columns)

# Taking only the categorical features that had the highest correlation to SalePrice
X_cat_train = cat_train[['Neighborhood_BrkSide', 'Neighborhood_Edwards', 'Neighborhood_NoRidge',
                         'Neighborhood_NridgHt', 'Neighborhood_OldTown',
                         'Neighborhood_StoneBr', 'Exterior 1st_CemntBd']]

# Doing the same for the final test data
X_cat_final = cat_test[['Neighborhood_BrkSide', 'Neighborhood_Edwards', 'Neighborhood_NoRidge',
                         'Neighborhood_NridgHt', 'Neighborhood_OldTown',
                         'Neighborhood_StoneBr', 'Exterior 1st_CemntBd']]

# Finally combining all the numerical and categorical data into one dataset
X = pd.concat([X_nums, X_cat_train], axis = 1, sort = False)
X_final = pd.concat([X_nums_final, X_cat_final], axis = 1, sort = False)

# Train-test-splitting the data for modeling and evaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# XGBoost parameters that will be gridsearched over
xgb_params = {
    'eta' : [0.1,0.3],
    'min_child_weight' : [1, 2, 3],
    'max_depth' : [3, 5 ,7, 9],
    'colsample_bytree' : [.5, .7, .9, 1]
}

# Initializing GridSearch over XGBoost
grid = GridSearchCV(
    XGBRegressor(),
    xgb_params,
    cv = 5,
    verbose = 1,
    return_train_score = True
)

# Fitting the model
grid.fit(X_train, y_train)
```

```python
# The best_params_ attribute shows the highest scoring 
# hyperparameters that GridSearch found after fitting
grid.best_params_
```

```python
# The best_score_ is the highest R2 score after fitting
# of all the hyperparameter combinations 
grid.best_score_
```

```python
# Using the best parameters, I made a list of predictions to compare to the test split
y_hat = grid.predict(X_test)

# Getting the RMSE of test split vs. predictions
mean_squared_error(y_test,y_hat) ** 0.5
```

```python
# Graphing true values vs. predicted values
plt.figure(figsize=(10,8), facecolor='w')
plt.title('XGBoost w/ FE: Actual Price vs. Predicted Price', fontsize = 20)
plt.ylabel('Actual Price', fontsize = 20)
plt.xlabel('Predicted Price', fontsize = 20)
plt.plot(np.unique(y_test), 
         np.poly1d(np.polyfit(y_test, y_hat, 1))(np.unique(y_test)),
         color = 'red')
plt.scatter(y_test, y_hat)
plt.text(x=320000,y=25000, s="RMSE: 21737.11", fontsize=20);
plt.savefig('../images/xgboost1.png', bbox_inches='tight')
```

_Without much effort, the XGBoost model obtained the lowest RMSE score compared to the other models that I had previously used in my project with the Ames, IA housing dataset.
