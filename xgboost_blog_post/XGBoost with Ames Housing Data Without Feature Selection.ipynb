# XGBoost with the Ames Housing Dataset Without Feature Engineering

### Using XGBoost with regression
_This notebook is designed to take cleaned data from the Ames, Iowa housing dataset and apply a linear regression model with XGBoost regression. This model will be using data that has not been feature engineered, but rather includes all the available features from the data._

```python
# Loading appropriate libraries
import xgboost
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from xgboost.sklearn import XGBRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)

%matplotlib inline
```

```python
# Loading in data from previously cleaned datasets
num_train = pd.read_csv('../datasets/num_train.csv')
cat_train = pd.read_csv('../datasets/cat_train.csv')
num_test = pd.read_csv('../datasets/num_test.csv')
cat_test = pd.read_csv('../datasets/cat_test.csv')
train = pd.read_csv('../datasets/train.csv')
test = pd.read_csv('../datasets/test.csv')

# Dropping target feature from train data
num_train = num_train.drop('SalePrice',axis=1)

# Applying standarization to numerical features
ss = StandardScaler()
X_nums = pd.DataFrame(ss.fit_transform(num_train), columns = num_train.columns)
X_nums_final = pd.DataFrame(ss.transform(num_test), columns = num_test.columns)

# Combining numerical and categorical features
X = pd.concat([X_nums, cat_train], axis = 1, sort = False)
X_final = pd.concat([X_nums_final, cat_test], axis = 1, sort = False)

# Extracting y from original train data
y = train['SalePrice']

# Train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# XGBoost parameters that will be gridsearched over
xgb_params = {
    'eta' : [0.1,0.3],
    'min_child_weight' : [1, 2, 3],
    'max_depth' : [3, 5 ,7, 9],
    'colsample_bytree' : [.5, .7, .9, 1]
}

# Initializing GridSearch over XGBoost
grid = GridSearchCV(
    XGBRegressor(),
    xgb_params,
    cv = 5,
    verbose = 1,
    return_train_score = True
)

# Fitting the model
grid.fit(X_train, y_train)
```

```python
# The best_params_ attribute shows the highest scoring 
# hyperparameters that GridSearch found after fitting
grid.best_params_
```

```python
# The best_score_ is the highest R2 score after fitting
# of all the hyperparameter combinations 
grid.best_score_
```

```python
# Using the best parameters, I made a list of predictions to compare to the test split
y_hat = grid.predict(X_test)

# Getting the RMSE of test split vs. predictions
mean_squared_error(y_test,y_hat) ** 0.5
```

```python
# Graphing true values vs. predicted values
plt.figure(figsize=(10,8), facecolor='w')
plt.title('XGBoost w/o FE: Actual Price vs. Predicted Price', fontsize = 20)
plt.ylabel('Actual Price', fontsize = 20)
plt.xlabel('Predicted Price', fontsize = 20)
plt.plot(np.unique(y_test), 
         np.poly1d(np.polyfit(y_test, y_hat, 1))(np.unique(y_test)),
         color = 'red')
plt.scatter(y_test, y_hat)
plt.text(x=320000,y=25000, s="RMSE: 22059.28", fontsize=20);
plt.savefig('../images/xgboost2.png', bbox_inches='tight')
```
