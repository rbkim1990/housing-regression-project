# Cleaning Numeric Features

_This notebook is dedicated to cleaning and selecting numerical features._

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split
```

_I imported the .csv files from the notebook that cleaned the categorical data (the features that were converted from categorical to numerical data)._

```python
train = pd.read_csv('../datasets/train_clean.csv')
test = pd.read_csv('../datasets/test_clean.csv')
train_add = pd.read_csv('../datasets/noncat_train.csv')
test_add = pd.read_csv('../datasets/noncat_test.csv')
other = pd.read_csv('../datasets/train.csv')
```

```python
train = train._get_numeric_data()
test = test._get_numeric_data()
```

```python
train.shape
```

```python
test.shape
```

_I dropped a few features that I thought would have no impact on the SalePrice._

```python
train.drop(['Id', 'PID', 'MS SubClass', 'Lot Frontage', 'Mo Sold'],axis=1,inplace=True)
test.drop(['Id', 'PID', 'MS SubClass', 'Lot Frontage', 'Mo Sold'],axis=1,inplace=True)
```

_Concatenated the DataFrames to get the complete numeric features that I wanted._

```python
train.shape
```

```python
test.shape
```

```python
train = pd.concat([train,train_add],axis=1,sort=False)
test = pd.concat([test,test_add],axis=1,sort=False)
```

```python
train.shape
```

```python
test.shape
```

_Saved the data into their respective .csv files._

```python
train.to_csv('../datasets/num_train.csv', index = False)
test.to_csv('../datasets/num_test.csv', index = False)
```

```python
train.columns[(train.corr()['SalePrice'] > 0.5)]
```

_At this point in the notebook, I decided to use the Lasso regression to zero out columns._

```python
cols = [index for index,col in enumerate(train.columns) if col != 'SalePrice']
X = train.iloc[0:,cols]
y = train['SalePrice']
X_final = test
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
lasso = LassoCV(n_alphas = 200)
lasso.fit(X_train, y_train)
lasso.score(X_test, y_test)
```

```python
not_zero = [col for col, val in enumerate(lasso.coef_) if val != 0]
```

```python
lasso.coef_
```

_The result of the Lasso regression is the above columns that were not zero-ed out. I will use this information going forward._

```python
df = X[X_train.columns[not_zero]]
```

```python
df['SalePrice'] = other['SalePrice']
```
